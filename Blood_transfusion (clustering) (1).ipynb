```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_aa6d53b46ca443dea5b8031d958bc35c.csv"
df=pd.read_csv(url)
df.head()
```

```python
X=df.drop(["Frequency (times)","Time (months)","whether he/she donated blood in March 2007"],axis=1)
X.head()
```

```python
y=df[["whether he/she donated blood in March 2007"]]
y.head()
```

```python
import numpy as np
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
bandwidth = estimate_bandwidth(X,quantile=0.75, n_jobs=-1)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
labels=ms.fit_predict(X)
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)
print("The number of clusters is: ",n_clusters_)
print("The silhouette score is: ",silhouette_score(X, labels))
```

```python
label=pd.DataFrame(labels).rename(columns={0:"Label"})
```

```python
comparison=pd.concat([X,y,label],axis=1)
comparison.head()
```

```python
accuracy=pd.DataFrame(comparison["whether he/she donated blood in March 2007"]==comparison["Label"]).rename(columns={0:"Accurate"})
accuracy.head(10)
```

```python
len(accuracy[accuracy["Accurate"]==True])/len(comparison)
```

```python
indices=accuracy[accuracy["Accurate"]==True]
indices=indices.index.tolist()
```

```python
accurate=comparison.loc[indices,:]
accurate.head()
```

```python
len(accurate[accurate["Label"]==1])
```

```python
from sklearn.neighbors import NearestCentroid
clf = NearestCentroid()
clf.fit(X, labels)
clf.centroids_
```
